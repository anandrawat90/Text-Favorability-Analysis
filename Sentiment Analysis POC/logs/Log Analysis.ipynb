```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('TL_BI_cuDNNLSTM_L2_N196_GloVe_5f_final.csv',encoding='utf-8')
```

```python
data
```

```python
epoch = [x for x in range(10)]
```

```python
data.epoch = epoch
```

```python
print("Mean Validation Accuracy: %.2f%% Standard Deviation(+/- %.2f%%)" % (np.mean(data.val_acc)*100, np.std(data.val_acc)*100))
print("Mean Validation Loss: %.2f Standard Deviation(+/- %.2f)" % (np.mean(data.val_loss), np.std(data.val_loss)))
```

```python
%matplotlib inline
```

```python
acc_data = data[['acc','val_acc']].plot(kind = 'line',title='Traning Accuracy Vs Validation Accuracy',fontsize=18,grid=True)
fig = acc_data.get_figure()
fig.set_size_inches(18.5,10.5)
fig.savefig('Accuracy.png',format='png',dpi=100)
```

```python
loss_data = data[['loss','val_loss']].plot(kind = 'line',title='Traning Loss Vs Validation Loss',fontsize=18,grid=True)
fig = loss_data.get_figure()
fig.set_size_inches(18.5,10.5)
fig.savefig('Loss.png',format='png',dpi=100)
```
