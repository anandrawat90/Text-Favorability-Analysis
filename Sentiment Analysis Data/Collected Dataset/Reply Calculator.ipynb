```python
import pandas as pd
import numpy as np
```

```python
org_tweet = pd.read_csv('WorldCupDraw_Cleansed_final.csv',encoding = 'utf-8')
```

```python
replies = pd.read_csv('Replies_WorldCupDraw_Cleansed_Finalized_2.csv',dtype= {"original_tweeter": object },encoding = 'utf-8')
```

```python
replies.shape
```

```python
org_tweet.shape
```

```python
dataset = pd.merge(org_tweet,replies,on='id',suffixes=('_org','_rep'))
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
stat_set = dataset.groupby('id')
```

```python
stat_set.head()
```

```python
count = dict()
for name, group in stat_set:
    pos_cnt = neg_cnt = 0
    for pred in group.Prediction:
        if pred == 0:
            neg_cnt += 1
        else:
            pos_cnt += 1
    count[name] = [pos_cnt,neg_cnt]
```

```python
len(count)
```

```python
org_tweet['num_positive_replies'] = np.zeros(org_tweet.shape[0])
org_tweet['num_negative_replies'] = np.zeros(org_tweet.shape[0])
```

```python
org_tweet.set_index('id',inplace=True)
```

```python
org_tweet.head()
```

```python
org_tweet.at[937247505208827904,"num_positive_replies"] = 1
```

```python
for key in count:
    org_tweet.at[key,"num_positive_replies"] = count[key][0]
    org_tweet.at[key,"num_negative_replies"] = count[key][1]
```

```python
org_tweet.shape
```

```python
org_tweet.head()
```

```python
org_tweet.num_positive_replies = org_tweet.num_positive_replies.astype('int32',copy=False)
org_tweet.num_negative_replies = org_tweet.num_negative_replies.astype('int32',copy=False)
```

```python
org_tweet.reset_index(inplace=True)
```

```python
org_tweet.to_csv(path_or_buf='WorldCupDraw Preprocessed Dataset.csv',encoding = 'utf-8',index=False)
```

```python
print('Average # Positive replies: %.2f , Average # Negative replies: %.2f '%(org_tweet.num_positive_replies.mean(),org_tweet.num_negative_replies.mean()))
```

```python
org_tweet['positive_percentage'] = np.zeros(len(count))
```

```python
for key in count:
    org_tweet.at[key,'positive_percentage'] = round(float((org_tweet.at[key,'num_positive_replies']/org_tweet.at[key,'reply_count'])*100),2)
```
